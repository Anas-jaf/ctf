```python
import requests

response = requests.get('http://45.33.98.250:1339/robots.txt')
```

```python
nr = response.text.replace('User-agent: *','')
nr = nr.split('Disallow: ')
nr = [i.replace('\n' , '') for i in nr if len(i.replace('\n' , '')) !=0  ]
nr
```

```python
for flag in nr : 
    if len(flag) !=0 :
        response = requests.get(f'http://45.33.98.250:1339/{flag}')
        if response.status_code == 200:
            print(flag)
            print(response.text)
```

```python
possible_passwords = [i.replace('/','').replace('.php','') for i in nr ]

possible_passwords
```
